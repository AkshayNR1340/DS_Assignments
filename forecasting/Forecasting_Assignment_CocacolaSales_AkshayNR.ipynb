# Advanced exponential method

# Holt's winter model

Holts winter exponential smoothing with multiplicative seasonality and additive trend

We select Holt's winter method only when we have both SEASONALITY and TREND

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing 
from sklearn.metrics import mean_squared_error
mse=[]
import warnings
warnings.filterwarnings('ignore')
```

```python
hwe_model_mul_add = ExponentialSmoothing(Train["Sales"],seasonal="mul",trend="add",
                                         seasonal_periods=4).fit() 
pred_hwe_mul_add = hwe_model_mul_add.predict(start = Test.index[0],end = Test.index[-1])
error=np.sqrt(mean_squared_error(Test.Sales,pred_hwe_mul_add))
mse.append(error)  
print((error))   
```

```python
Train.plot(legend=True,label='Train',figsize=(10,6))
Test.plot(legend=True,label='Test')
pred_hwe_mul_add.plot(legend=True,label='Predicted')
```

# Building ARIMA model on transformed data

```python
import statsmodels.graphics.tsaplots as tsa_plots
tsa_plots.plot_acf(coca1sqrt.Sales,lags=12)
plt.show()
```

```python
tsa_plots.plot_pacf(coca1sqrt.Sales,lags=12)
plt.show()
```

```python
import itertools
```

```python
p=range(0,6)
q=range(0,6)
d=range(0,3)
pdq_combination=list(itertools.product(p,d,q))
```

```python
len(pdq_combination)
order=[]
mse=[]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
for pdq in pdq_combination:
        model=sm.tsa.arima.ARIMA(Train,order=pdq).fit()
        pred=model.predict(start=len(Train),end=(len(coca1sqrt)-1))
        Error=np.sqrt(mean_squared_error(Test,pred))
        order.append(pdq)
        mse.append(Error)  
        print('Best ARIMA%s RMSE=%.2f' % (pdq,Error))   
```

(2,2,3) got less error
