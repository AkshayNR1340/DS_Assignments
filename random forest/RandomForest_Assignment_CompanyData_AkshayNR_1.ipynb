Splitting the default/normal data into train and test

```python
x=cdata_label.iloc[:,1:]
y=cdata_label.iloc[:,0]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

Standardising and splitting the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
xscaled=scaler.fit_transform(x)
xscaled
```

```python
df=pd.DataFrame(xscaled,columns=cdata.columns[1:])
df
```

```python
X=df.iloc[:,:]
Y=cdata_label.iloc[:,0]
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=40)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

# Building model using Random Forest Classification on default data

```python
kfold = KFold(n_splits=10, random_state=7,shuffle=True)
model = RandomForestClassifier(n_estimators=700, max_features=5,criterion='gini')
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```

```python
#Model on default/normal data splitted
model.fit(x_train,y_train)
```

```python
y_pred_test=model.predict(x_test)
acc=accuracy_score(y_test, y_pred_test)*100
print("Accuracy=", acc)
print(confusion_matrix(y_test, y_pred_test))
```

# Building model using RandomForest Classification on Standardised data

```python
kfold = KFold(n_splits=10, random_state=7,shuffle=True)
stdmodel = RandomForestClassifier(n_estimators=700, max_features=5,criterion='entropy')
results = cross_val_score(stdmodel,X ,Y, cv=kfold)
print(results.mean())
```

```python
stdmodel.fit(X_train,Y_train)
```

```python
Y_pred_test=stdmodel.predict(X_test)
acc=accuracy_score(Y_test, Y_pred_test)*100
print("Accuracy=", acc)
print(confusion_matrix(Y_test, Y_pred_test))
```

Here on standardised model we got better accuracy than model created using normal/default data.
