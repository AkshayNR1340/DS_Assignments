```python
dfop10r['Reviews'] = dfop10r['Reviews'].fillna('').apply(str)
```

```python
def removed_pattern(input_txt1, pattern1):
    r1=re.findall(pattern1,input_txt1)
    for word in r1:
        input_txt1=re.sub(word,"",input_txt1)
    return input_txt1
```

```python
def clean_text(text):
    '''make text lowercase,remove text in square brackets,remove punchuation and remove containing numbers.'''
    text=re.sub("[0-9" "]+"," ",text)
    text=re.sub('\[.*?\]','',text)
    text=re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text=re.sub('\w*\d\w*','',text)
    text=re.sub('[''""...]','',text)
    text=text.lower()
    return text
clean=lambda x: clean_text(x)
```

```python
dfop10r['cleaned_reviews']=dfop10r.Reviews.apply(clean)
```

# N-gram

```python
from sklearn.feature_extraction.text import CountVectorizer
#bi-gram
def get_top_n2_words(corpus, n=None):
    vec1=CountVectorizer(ngram_range=(2,2), #for tri-gram, put ngram_range=(3,3)
          max_features=2000).fit(corpus)               
    bag_of_words=vec1.transform(corpus)
    sum_words=bag_of_words.sum(axis=0)
    words_freq=[(word,sum_words[0, idx]) for word, idx in
               vec1.vocabulary_.items()]
    words_freq=sorted(words_freq, key=lambda x: x[1],
            reverse=True)
    return words_freq[:n]
```

```python
top2_words=get_top_n2_words(dfop10r['cleaned_reviews'],n=200) #top 200
top2_df=pd.DataFrame(top2_words)
top2_df.columns=["Bi-gram","Freq"]
top2_df.head()
```

```python
#Bi-gram plot
top20_bigram=top2_df.iloc[0:20,:]
fig=plt.figure(figsize=(10,5))
plot=sns.barplot(x=top20_bigram["Bi-gram"],y=top20_bigram["Freq"])
plot.set_xticklabels(rotation=45,labels=top20_bigram["Bi-gram"]);
```

```python
#tri-gram
def get_top_n3_words(corpus, n=None):
    vec1=CountVectorizer(ngram_range=(3,3),
          max_features=2000).fit(corpus)               
    bag_of_words=vec1.transform(corpus)
    sum_words=bag_of_words.sum(axis=0)
    words_freq=[(word,sum_words[0, idx]) for word, idx in
               vec1.vocabulary_.items()]
    words_freq=sorted(words_freq, key=lambda x: x[1],
            reverse=True)
    return words_freq[:n]
```

```python
top3_words=get_top_n3_words(dfop10r['cleaned_reviews'],n=200) 
top3_df=pd.DataFrame(top3_words)
top3_df.columns=["Tri-gram","Freq"]
top3_df.head()
```

```python
#tri-gram plot
top20_trigram=top3_df.iloc[0:20,:]
fig=plt.figure(figsize=(10,5))
plot1=sns.barplot(x=top20_trigram["Tri-gram"],y=top20_trigram["Freq"])
plot1.set_xticklabels(rotation=45,labels=top20_trigram["Tri-gram"]);
```

# splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(bow1,dfop10r['ratings'],test_size=0.25,random_state=35)
```

# Building Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
y_pred_train= model.predict(x_train)
acc = accuracy_score(y_train, y_pred_train) * 100
print("Accuracy =", acc)
print(confusion_matrix(y_train, y_pred_train))
```

```python
y_pred_test=model.predict(x_test)
acc=accuracy_score(y_test, y_pred_test)*100
print("Accuracy=", acc)
print(confusion_matrix(y_test, y_pred_test))
```

We got good accuracy

# Building Naive Bayes Model without using TFIDF

```python
dfop10rC=dfop10r.drop(['Reviews'],axis=1)
dfop10rC
```

```python
def split_into_words(i):
    return (i.split(" "))
```

```python
Reviews_train,Reviews_test= train_test_split(dfop10rC,test_size=0.3)
```

```python
Reviews_train
```

```python
Reviews_test
```

```python
#preparing reviews texts into word count matrix format
reviews_bow=CountVectorizer(analyzer=split_into_words).fit(dfop10rC.cleaned_reviews)
```

```python
#for all reviews
all_reviews_matrix=reviews_bow.transform(dfop10rC.cleaned_reviews)
all_reviews_matrix.shape
```

```python
#for training reviews
train_reviews_matrix=reviews_bow.transform(Reviews_train.cleaned_reviews)
train_reviews_matrix.shape
```

```python
#for testing reviews
test_reviews_matrix=reviews_bow.transform(Reviews_test.cleaned_reviews)
test_reviews_matrix.shape
```

```python
################ without TFIDF matrices########## 
#preparing a naive bayes model on training data set

from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
```

# Multinomial Naive Bayes Model

```python
classifier_mb=MB()
classifier_mb.fit(train_reviews_matrix,Reviews_train.ratings)
train_pred_m=classifier_mb.predict(train_reviews_matrix)
accuracy_train_m=np.mean(train_pred_m==Reviews_train.ratings)

test_pred_m=classifier_mb.predict(test_reviews_matrix)
accuracy_test_m=np.mean(test_pred_m==Reviews_test.ratings)
```

```python
accuracy_train_m
```

```python
accuracy_test_m
```

We got good accuracy and is almost same as LogisticRegresion test accuracy.

# Gaussian Naive Bayes Model

```python
classifier_gb=GB()
classifier_gb.fit(train_reviews_matrix.toarray(),Reviews_train.ratings.values)    #we need to convert tfidf into array format which is compatible
train_pred_g=classifier_gb.predict(train_reviews_matrix.toarray())
accuracy_train_g=np.mean(train_pred_m==Reviews_train.ratings)

test_pred_g=classifier_gb.predict(test_reviews_matrix.toarray())
accuracy_test_g=np.mean(test_pred_g==Reviews_test.ratings)
```

```python
accuracy_train_g
```

```python
accuracy_test_g
```

We got good accuracy here too but accuracy of MB is better than GB. And accuracy of Logisticregression Model was better tha Naive Bayes Models.
