```python
model=smf.ols("delivery_time~sorting_time",data=data).fit()
```

```python
model.summary()
```

```python
model1=smf.ols("delivery_time~np.log(sorting_time)",data=data).fit()
```

```python
model1.summary()
```

```python
model2=smf.ols("np.log(delivery_time)~sorting_time",data=data).fit()
```

```python
model2.summary()
```

```python
model3=smf.ols("np.sqrt(delivery_time)~(sorting_time)",data=data).fit()
```

```python
model3.summary()
```

```python
data['sorting_time_sq']=data.sorting_time**2
model4=smf.ols('np.log(delivery_time)~sorting_time+sorting_time_sq',data=data).fit()
```
